## Customer_purchase_behaviour_KNN_classification_model

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/PankajVispute/I-phone-purchase-project--Prediction-with-KNN-Classification/main/iphone_purchase_records.csv")
```

```python
df
```

```python
gender_encoded=pd.get_dummies(df['Gender'])
```

```python
gender_encoded
```

```python
gender_encoded=pd.get_dummies(df['Gender'],drop_first=True)
```

```python
gender_encoded
```

```python
df=pd.concat([df,gender_encoded],axis=1)
```

```python
df
```

```python
x=df[['Age','Salary','Male']].to_numpy()
```

```python
y=df['Purchase Iphone'].to_numpy()
```

```python
x
```

```python
import sklearn as sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
y_train
```

```python
y_test
```

```python
len(x_train)
```

```python
len(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)

```

```python
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k=9
knn=KNeighborsClassifier(n_neighbors=k)
```

```python
knn.fit(x_train,y_train)
```

```python
y_predicted=knn.predict(x_test)
y_predicted
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
model_accuracy=accuracy_score(y_test,y_predicted)
model_accuracy
```

### Saving and Loading model using Pickle module

```python
import pickle
with open('model_pickle.pkl','wb')as f:
    pickle.dump(knn,f)
```

```python
with open('model_pickle.pkl','rb')as f:
    model_pickle=pickle.load(f)
```
